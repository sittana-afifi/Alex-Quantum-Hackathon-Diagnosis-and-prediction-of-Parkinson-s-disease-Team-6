```python
# Advanced Neural Network for Parkinson's Disease Classification
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, regularizers
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.preprocessing import StandardScaler, RobustScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
from sklearn.utils.class_weight import compute_class_weight
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
from imblearn.over_sampling import SMOTE
%load_ext tensorboard

# Load data from CSV file
data = pd.read_csv('pd_speech_features.csv', header=1)

# Data preprocessing
X = data.drop(['class'], axis=1)
y = data['class']

# Check for missing values and handle them if any
if X.isnull().sum().sum() > 0:
    for column in X.columns:
        if X[column].isnull().sum() > 0:
            X[column].fillna(X[column].mean(), inplace=True)

# Handle class imbalance using SMOTE
print(f"Class distribution before SMOTE: {np.bincount(y)}")
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)
print(f"Class distribution after SMOTE: {np.bincount(y_resampled)}")

# Feature scaling - using RobustScaler which is less sensitive to outliers
scaler = RobustScaler()
X_scaled = scaler.fit_transform(X_resampled)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_resampled, test_size=0.15, random_state=42, stratify=y_resampled
)

# Further split training set for validation
X_train, X_val, y_train, y_val = train_test_split(
    X_train, y_train, test_size=0.15, random_state=42, stratify=y_train
)

print(f"Training set size: {X_train.shape}")
print(f"Validation set size: {X_val.shape}")
print(f"Testing set size: {X_test.shape}")

# Convert to tensors
X_train_tensor = tf.convert_to_tensor(X_train, dtype=tf.float32)
X_val_tensor = tf.convert_to_tensor(X_val, dtype=tf.float32)
X_test_tensor = tf.convert_to_tensor(X_test, dtype=tf.float32)
y_train_tensor = tf.convert_to_tensor(y_train, dtype=tf.float32)
y_val_tensor = tf.convert_to_tensor(y_val, dtype=tf.float32)
y_test_tensor = tf.convert_to_tensor(y_test, dtype=tf.float32)

# Save processed data
processed_data = {
    'X_train': X_train,
    'X_val': X_val,
    'X_test': X_test,
    'y_train': y_train,
    'y_val': y_val,
    'y_test': y_test,
    'scaler': scaler
}

joblib.dump(processed_data, 'parkinsons_advanced_processed_data.pkl')

# Calculate class weights for imbalanced data
class_weights = compute_class_weight('balanced', classes=np.unique(y_train), y=y_train)
class_weights = dict(enumerate(class_weights))
print(f"Class weights: {class_weights}")

# Build advanced neural network
def create_advanced_nn(input_dim):
    model = keras.Sequential([
        # Input layer
        layers.Dense(128, activation='relu', input_shape=(input_dim,),
                    kernel_regularizer=regularizers.l2(0.001)),
        layers.BatchNormalization(),
        layers.Dropout(0.4),

        # Hidden layers
        layers.Dense(96, activation='relu',
                    kernel_regularizer=regularizers.l2(0.001)),
        layers.BatchNormalization(),
        layers.Dropout(0.4),

        layers.Dense(64, activation='relu',
                    kernel_regularizer=regularizers.l2(0.001)),
        layers.BatchNormalization(),
        layers.Dropout(0.3),

        layers.Dense(32, activation='relu',
                    kernel_regularizer=regularizers.l2(0.001)),
        layers.BatchNormalization(),
        layers.Dropout(0.3),

        layers.Dense(16, activation='relu',
                    kernel_regularizer=regularizers.l2(0.001)),
        layers.BatchNormalization(),
        layers.Dropout(0.2),

        # Output layer
        layers.Dense(1, activation='sigmoid')
    ])

    return model

# Create and compile model
advanced_model = create_advanced_nn(X_train.shape[1])
advanced_model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    loss='binary_crossentropy',
    metrics=['accuracy',
             keras.metrics.Precision(name='precision'),
             keras.metrics.Recall(name='recall')]
)

# Display model summary
advanced_model.summary()

# Callbacks
callbacks = [
    EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=7, min_lr=1e-7),
    ModelCheckpoint('best_model.h5', monitor='val_accuracy', save_best_only=True)
]

# Train the model
history = advanced_model.fit(
    X_train_tensor, y_train_tensor,
    epochs=200,
    batch_size=16,
    validation_data=(X_val_tensor, y_val_tensor),
    class_weight=class_weights,
    callbacks=callbacks,
    verbose=1
)

# Load the best model
advanced_model = keras.models.load_model('best_model.h5')

# Plot training history
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 3, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1, 3, 3)
plt.plot(history.history['precision'], label='Training Precision')
plt.plot(history.history['val_precision'], label='Validation Precision')
plt.plot(history.history['recall'], label='Training Recall')
plt.plot(history.history['val_recall'], label='Validation Recall')
plt.title('Precision and Recall')
plt.xlabel('Epoch')
plt.ylabel('Score')
plt.legend()

plt.tight_layout()
plt.savefig('training_history.png', dpi=300)
plt.show()

# Make predictions
y_pred_probs = advanced_model.predict(X_test_tensor)
y_pred = (y_pred_probs > 0.5).astype(int).flatten()

# Calculate metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)

print("Neural Network Performance:")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
print("Confusion Matrix:")
print(cm)
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Plot confusion matrix
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('NN Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.savefig('advanced_confusion_matrix.png', dpi=300)
plt.show()

# If it need more improvement, let's try ensemble methods
if accuracy < 0.95:
    print("Accuracy need enhancement. Trying ensemble methods...")

    # Create an ensemble of models
    from sklearn.ensemble import VotingClassifier, StackingClassifier
    from sklearn.svm import SVC
    from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
    from xgboost import XGBClassifier
    from sklearn.linear_model import LogisticRegression

    # Define base models
    estimators = [
        ('rf', RandomForestClassifier(n_estimators=200, random_state=42)),
        ('gb', GradientBoostingClassifier(n_estimators=200, random_state=42)),
        ('xgb', XGBClassifier(n_estimators=200, random_state=42, use_label_encoder=False, eval_metric='logloss')),
        ('svc', SVC(probability=True, random_state=42))
    ]

    # Create voting classifier
    voting_clf = VotingClassifier(estimators=estimators, voting='soft')
    voting_clf.fit(X_train, y_train)

    # Make predictions with ensemble
    y_pred_ensemble = voting_clf.predict(X_test)

    # Calculate metrics for ensemble
    accuracy_ensemble = accuracy_score(y_test, y_pred_ensemble)
    precision_ensemble = precision_score(y_test, y_pred_ensemble)
    recall_ensemble = recall_score(y_test, y_pred_ensemble)
    f1_ensemble = f1_score(y_test, y_pred_ensemble)

    print("Ensemble Model Performance:")
    print(f"Accuracy: {accuracy_ensemble:.4f}")
    print(f"Precision: {precision_ensemble:.4f}")
    print(f"Recall: {recall_ensemble:.4f}")
    print(f"F1 Score: {f1_ensemble:.4f}")

    # If ensemble performs better, use it
    if accuracy_ensemble > accuracy:
        accuracy = accuracy_ensemble
        precision = precision_ensemble
        recall = recall_ensemble
        f1 = f1_ensemble
        y_pred = y_pred_ensemble
        cm = confusion_matrix(y_test, y_pred_ensemble)

        print("Using ensemble model as it performed better.")

# Save advanced metrics
advanced_metrics = {
    'accuracy': accuracy,
    'precision': precision,
    'recall': recall,
    'f1': f1,
    'confusion_matrix': cm
}

joblib.dump(advanced_metrics, 'advanced_model_metrics.pkl')

# If we still need improvement, try feature engineering
if accuracy < 0.95:
    print("Still need enhancement. Trying feature engineering...")

    # Create polynomial features
    from sklearn.preprocessing import PolynomialFeatures
    from sklearn.feature_selection import SelectKBest, f_classif

    # Create polynomial features
    poly = PolynomialFeatures(degree=2, interaction_only=True, include_bias=False)
    X_poly = poly.fit_transform(X_resampled)

    # Select best features
    selector = SelectKBest(score_func=f_classif, k=min(50, X_poly.shape[1]))
    X_selected = selector.fit_transform(X_poly, y_resampled)

    print(f"Original features: {X_resampled.shape[1]}")
    print(f"After polynomial features and selection: {X_selected.shape[1]}")

    # Split the enhanced dataset
    X_train_fe, X_test_fe, y_train_fe, y_test_fe = train_test_split(
        X_selected, y_resampled, test_size=0.15, random_state=42, stratify=y_resampled
    )

    # Train a model on enhanced features
    model_fe = create_advanced_nn(X_train_fe.shape[1])
    model_fe.compile(
        optimizer=keras.optimizers.Adam(learning_rate=0.001),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )

    # Train with early stopping
    history_fe = model_fe.fit(
        X_train_fe, y_train_fe,
        epochs=100,
        batch_size=16,
        validation_split=0.15,
        callbacks=[EarlyStopping(patience=10, restore_best_weights=True)],
        verbose=0
    )

    # Evaluate
    y_pred_fe = (model_fe.predict(X_test_fe) > 0.5).astype(int).flatten()
    accuracy_fe = accuracy_score(y_test_fe, y_pred_fe)

    print(f"Accuracy with feature engineering: {accuracy_fe:.4f}")

    # If feature engineering helps, use it
    if accuracy_fe > accuracy:
        accuracy = accuracy_fe
        y_pred = y_pred_fe
        print("Using feature engineering as it improved performance.")

# Final evaluation
if accuracy >= 0.95:
    print(f"SUCCESS: Achieved {accuracy*100:.2f}% accuracy!")
else:
    print(f"FINAL RESULT: {accuracy*100:.2f}% accuracy")

# Plot feature importance
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=200, random_state=42)
rf.fit(X_train, y_train)

importances = rf.feature_importances_
feature_names = data.drop(['class'], axis=1).columns
indices = np.argsort(importances)[::-1]

plt.figure(figsize=(12, 8))
plt.title("Feature Importance")
plt.bar(range(15), importances[indices][:15])
plt.xticks(range(15), [feature_names[i] for i in indices[:15]], rotation=45, ha='right')
plt.tight_layout()
plt.savefig('feature_importance_advanced.png', dpi=300)
plt.show()

# Compare with previous models if available
try:
    classical_metrics = joblib.load('classical_model_metrics.pkl')
    #quantum_metrics = joblib.load('quantum_model_metrics.pkl')

    # Create comparison plot
    models = ['Classical NN', 'Quantum', 'Advanced NN']
    accuracies = [classical_metrics['accuracy'], quantum_metrics['accuracy'], accuracy]
    precisions = [classical_metrics['precision'], quantum_metrics['precision'], precision]
    recalls = [classical_metrics['recall'], quantum_metrics['recall'], recall]
    f1_scores = [classical_metrics['f1'], quantum_metrics['f1'], f1]

    x = np.arange(len(models))
    width = 0.2

    plt.figure(figsize=(12, 6))
    plt.bar(x - 1.5*width, accuracies, width, label='Accuracy')
    plt.bar(x - 0.5*width, precisions, width, label='Precision')
    plt.bar(x + 0.5*width, recalls, width, label='Recall')
    plt.bar(x + 1.5*width, f1_scores, width, label='F1 Score')

    plt.xlabel('Models')
    plt.ylabel('Scores')
    plt.title('Model Comparison')
    plt.xticks(x, models)
    plt.legend()
    plt.tight_layout()
    plt.savefig('all_models_comparison.png', dpi=300)
    plt.show()

except FileNotFoundError as e:
    print(f"Comparison not available: {e}")
```
